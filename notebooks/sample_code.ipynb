```python
import sys
import torch
import tiktoken
sys.path.append('../')
from modular.AssignWeights import model_config,load_weights_into_gpt
from modular.GPTArchitecture.GPTModel import GPTModel
from modular.LoadGPT2Weights import download_and_load_gpt2
from modular.GPTArchitecture.TextGeneration import token_ids_to_text,text_to_token_ids,generate
```

```python
tokenizer = tiktoken.get_encoding('gpt2')
```

Old config we used before using OpenAI weights (we set context_length = 256 for faster computations and qkv_bias=False to exclude the bias term when initializing weight matrices.)

```python
GPT_CONFIG_124M = {
 "vocab_size": 50257,      # Vocabulary size
 "context_length": 256,   # Context length
 "emb_dim": 768,           # Embedding dimension
 "n_heads": 12,            # Number of attention heads
 "n_layers": 12,           # Number of layers
 "drop_rate": 0.1,         # Dropout rate
 "qkv_bias": False         # Query-Key-Value bias
}
```

```python
NEW_CONFIG = model_config(GPT_CONFIG_124M)
gpt = GPTModel(NEW_CONFIG)
gpt.eval()
```

```python
settings,params = download_and_load_gpt2(model_size="124M",models_dir="gpt2")
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
load_weights_into_gpt(gpt,params)
gpt.to(device)
```

```python
torch.manual_seed(123)

token_ids = generate(
    model=gpt,
    idx=text_to_token_ids("life is", tokenizer).to(device),
    max_new_tokens=25,
    context_size=NEW_CONFIG["context_length"],
    top_k=50,
    temperature=1.5
)

print("Output text:\n", token_ids_to_text(token_ids, tokenizer))
```
